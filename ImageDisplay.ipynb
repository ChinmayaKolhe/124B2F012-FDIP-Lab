```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_path = "nature.jpg"

if not os.path.exists(image_path):
    print("Image not found.")
    exit()


image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

brightness_value = 50
image_bright = cv2.convertScaleAbs(image, alpha=1, beta=brightness_value)
image_bright_rgb = cv2.cvtColor(image_bright, cv2.COLOR_BGR2RGB)


_, image_thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)


plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Color (RGB)")
plt.axis('off')


plt.subplot(2, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(image_bright_rgb)
plt.title(f"Brightness +{brightness_value}")
plt.axis('off')


plt.subplot(2, 2, 4)
plt.imshow(image_thresh, cmap='gray')
plt.title("Thresholded (Binary)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
