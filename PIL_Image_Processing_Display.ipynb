```python
from PIL import Image, ImageDraw, ImageFont, ImageEnhance
from IPython.display import display
```

```python
image = Image.open("nature.jpg")
display(image)
```

```python
print("Size:", image.size)
print("Format:", image.format)
print("Mode:", image.mode)
```

```python
resized_image = image.resize((300, 200))
display(resized_image)
```

```python
cropped_image = image.crop((50, 50, 250, 200))
display(cropped_image)
```

```python
rotated_image = image.rotate(90)
display(rotated_image)

flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
display(flipped_image)
```

```python

text_img = image.copy()
draw = ImageDraw.Draw(text_img)
font = ImageFont.truetype("arial.ttf", 20)
draw.text((50, 50), "Hello Nature!", fill=(255, 255, 255), font=font)
display(text_img)
```

```python

bright_enhancer = ImageEnhance.Brightness(image)
bright_image = bright_enhancer.enhance(1.5)

contrast_enhancer = ImageEnhance.Contrast(bright_image)
contrast_image = contrast_enhancer.enhance(1.2)

display(contrast_image)
```

```python
contrast_image.save("processed_nature.png", format="PNG")
```
